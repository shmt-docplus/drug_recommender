```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os

# update options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

read the data 

```python
drug_records=pd.read_csv("50-70krecords_updated_11_09.zip",compression="zip",header=0)
```

```python
drug_records.head()
```

```python
drug_records.drop(['Unnamed: 0', 'S.no'],axis=1,inplace=True)
```

```python
droplist=['Quantity 6', 'Unit 6', 'Ingrediant 7', 'Quantity 7', 'Unit 7','Ingrediant 8', 'Quantity 8', 'Unit 8', 'How medicine works', 'Quick Tips']
```

```python
drug_records.drop(droplist, axis=1, inplace=True)
```

```python
drug_records.head()
```

```python
drug_types=drug_records.dtypes.reset_index()
drug_types.columns=["count","column_type"]
drug_types.groupby("column_type").aggregate("count").reset_index()
```
